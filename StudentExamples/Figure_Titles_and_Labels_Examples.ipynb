```python
from matplotlib import pyplot as plt
```

```python
some_data = [3, 1, 2, 3, 4, 2, 1, 2, 3, 2, 1, 2, 3, 3, 4, 4, 5, 2, 1, 2, 3, 4]
```

```python
fig0, ax0 = plt.subplots(figsize = (8, 8))
ax0.hist(some_data)
ax0.set_title("Histogram of Some Data") # apply set_title() to the axis you want.
ax0.set_xlabel("Some Data") # apply set_xlabel() to the axis to label the x axis of the plot
ax0.set_ylabel("Frequencies") # apply set_ylabel() to the axis to label the y axis of the plot
plt.show()
```

```python
fig1, ax1 = plt.subplots(1, 2, figsize = (8, 8))
ax1[0].hist(some_data, color = 'red')
ax1[0].set_title("Histogram of Some Data ax1[0]") # apply set_title() to the axis you want.
ax1[0].set_xlabel("Some Data") # apply set_xlabel() to the axis to label the x axis of the plot
ax1[0].set_ylabel("Frequencies") # apply set_ylabel() to the axis to label the y axis of the plot
ax1[1].hist(some_data, color = 'blue')
ax1[1].set_title("Histogram of Some Data  ax1[1]") # apply set_title() to the axis you want.
ax1[1].set_xlabel("Some Data") # apply set_xlabel() to the axis to label the x axis of the plot
ax1[1].set_ylabel("Frequencies") # apply set_ylabel() to the axis to label the y axis of the plot
plt.show()
```
