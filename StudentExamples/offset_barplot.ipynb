```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
champ_df = pd.DataFrame({'champ': ['cait', 'heim', 'jinx'], 'rate_pre': [.10, .11, .09], 'rate_post': [.13, .06, .08]})
```

```python
fig0, ax0 = plt.subplots(figsize = (10, 10))
ax0.bar(champ_df['champ'], champ_df['rate_pre'], alpha = .8, color = 'r', label = 'Pre Pickrate')
ax0.bar(champ_df['champ'], champ_df['rate_post'], alpha = .8, color = 'b', label = 'Post Pickrate')
plt.show()
```

```python
fig1, ax1 = plt.subplots(figsize = (10, 10))
x = [1, 2.25, 3.5]
ax1.bar([x-.25 for x in x], champ_df['rate_pre'], alpha = .8, color = 'r', label = 'Pre Pickrate', width = .5)
ax1.bar([x+.25 for x in x], champ_df['rate_post'], alpha = .8, color = 'b', label = 'Post Pickrate', width = .5)
ax1.set_xticks(x)
ax1.set_xticklabels(champ_df['champ'])
ax1.legend()
plt.show()
```
