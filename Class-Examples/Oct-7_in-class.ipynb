```python
import sqlite3
import pandas as pd
```

```python
loc_db = sqlite3.connect(':memory:')
```

```python
loc_db
```

```python
loc_cursor = loc_db.cursor()
```

```python
loc_cursor
```

```python
q_create_table = '''
CREATE TABLE IF NOT EXISTS book(id INTEGER PRIMARY KEY,
                                title TEXT,
                                author TEXT,
                                price INTEGER,
                                year TEXT)
'''
```

```python
loc_cursor.execute(q_create_table)
loc_db.commit()
```

```python
q_add_books = '''
INSERT INTO book (id, title, author, price, year)
VALUES
    (5, "Hitchhiker's Guide to the Galaxy", "Douglas Adams", 12.0, 1999),
    (6, "Cat's Craddle", 'Kurt Vonnegut', 20.25, 1963),
    (7, "Catch 22", "A Guy", 30000, 2070)
'''
```

```python
loc_cursor.execute(q_add_books)
loc_db.commit()
```

```python
q_select_book = '''
SELECT * from book
limit 3
'''
```

```python
book_ids = loc_cursor.execute(q_select_book).fetchall()
loc_db.commit()
```

```python
q_join_tracks = '''
SELECT a.name as Artist,
       title as Album,
       t.name as Song
FROM albums r
INNER JOIN artists a
ON (r.artistid = a.artistid)
INNER JOIN tracks t 
ON (r.albumid = t.albumid)
ORDER by a.name, title
'''
```

```python
all_track_info_df = pd.read_sql_query(q_join_tracks, chinook_db)
```
