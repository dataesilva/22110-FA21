```python
def get_covid_raffle_data(urls):
    df = pd.DataFrame()
    for url in urls:
        page = requests.request("GET", url)
        if page.status_code != 200:
            raise Exception(page.status_code, page.text)
            print(url)
        else:
            list_dfs = pd.read_html(page.text)
            data_chunk = list_dfs[0]
            df = df.append(data_chunk)
    return df
```

```python
urls = ['https://www.kent.edu/hr/news/employee-vaccination-drawing-week-4-winners',
       'https://www.kent.edu/hr/news/employee-vaccination-drawing-week-3-winners',
       'https://www.kent.edu/hr/faculty-staff-news/news/week-5-vaccination-winners']
```

```python
covid_winners_2 = get_covid_raffle_data(urls)
```
