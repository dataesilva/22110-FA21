```python
montero_track_data = pd.read_csv('Montero_Track_Data.csv')
equal_track_data = pd.read_csv('Equal_Track_Data.csv')
montero_audio_features = pd.read_csv('Montero_Audio_Features.csv')
equal_audio_features = pd.read_csv('Equal_Audio_Features.csv')
```

```python
fig1, ax1 = plt.subplots(1, 1, figsize = (10, 10))
m_offset = np.random.randn(len(montero_track_data.index)) * .02
e_offset = np.random.randn(len(equal_track_data.index)) * .02
m_x = [0] * len(montero_track_data.index)
e_x = [1] * len(equal_track_data.index)
m_jitter = m_x + m_offset
e_jitter = e_x + e_offset
ax1.scatter(y = montero_track_data['duration_ms'], x = m_jitter, color = 'blue', alpha = .5)
ax1.scatter(y = equal_track_data['duration_ms'], x = e_jitter, color = 'red', alpha = .5)
plt.show()
```

```python
fig2, ax2 = plt.subplots(1, 1, figsize = (10, 10))
m_offset = np.random.randn(len(montero_audio_features.index)) * .02
e_offset = np.random.randn(len(equal_audio_features.index)) * .02
m_x = [.2] * len(montero_audio_features.index)
e_x = [.8] * len(equal_audio_features.index)
m_jitter = m_x + m_offset
e_jitter = e_x + e_offset
ax2.scatter(y = montero_audio_features['danceability'], x = m_jitter, color = 'blue', alpha = .5)
ax2.scatter(y = equal_audio_features['danceability'], x = e_jitter, color = 'red', alpha = .5)
ax2.set_xlim(-.2, 1.2)
ax2.set_xticks([.2, .8])
ax2.set_xticklabels(['Montero', 'Equal'])
ax2.plot(.2, stats.describe(montero_audio_features['danceability']).mean, 'ko')
ax2.annotate('mean', (.2, stats.describe(montero_audio_features['danceability']).mean+.005))
ax2.plot(.8, stats.describe(equal_audio_features['danceability']).mean, 'ko')
plt.show()
```
